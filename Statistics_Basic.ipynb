```python
df.corr()
```

```python
from scipy import stats
#(corr,p-value)
r , p=stats.pearsonr(df.date,df.year)
print(round(r,4))
print(round(p))
```

```python
r , p=stats.pearsonr(df.date,df.id)
print(round(r,4))
print(round(p))
```

```python
r , p=stats.pearsonr(df.id,df.year)
print(round(r,4))
print(round(p))
```
